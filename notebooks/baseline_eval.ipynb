```python
import pandas as pd
from pathlib import Path
```

```python
input_path = Path('../reports/baseline_evaluation.csv')
```

```python
data = pd.read_csv(input_path)
data.head()
```

```python
SCORE_COLUMNS = [
    "clarity_score",
    "correctness_score",
    "exam_usefulness_score",
    "conciseness_score",
    "overall_score",
]
```

```python
avg_scores = data[SCORE_COLUMNS].mean()
topic_scores = data.groupby("topic")[SCORE_COLUMNS].mean()
```
