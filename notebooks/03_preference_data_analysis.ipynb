```python
import pandas as pd
```

```python
pref = pd.read_csv('../data/preferences/preference_labels_100.csv')
pref.head(5)
```

```python
pref['label'].value_counts()
```

```python
pref['chosen'] = ""
pref['rejected'] = ""
```

```python
for i, row in pref.iterrows():
    if pref.loc[i, 'label'] == 'A':
        pref.loc[i, 'chosen'] == pref.loc[i, 'answer_a']
        pref.loc[i, 'rejected'] == pref.loc[i, 'answer_b']
    elif pref.loc[i, 'label'] == 'B':
        pref.loc[i, 'chosen'] == pref.loc[i, 'answer_b']
        pref.loc[i, 'rejected'] == pref.loc[i, 'answer_a']
```
